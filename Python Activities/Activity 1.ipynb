## 3. Filter a DataFrame to include only rows where the date falls within a specified range.

```python
import pandas as pd

# Sample data
data = {
    'transaction_id': [101, 102, 103, 104, 105],
    'transaction_date': ['2024-06-01', '2024-06-15', '2024-07-01', '2024-07-15', '2024-08-01'],
    'amount': [100, 200, 150, 300, 250]
}

# Creating DataFrame
df = pd.DataFrame(data)
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
start_date = '2024-06-01'
end_date = '2024-07-01'
mask = (df['transaction_date'] >= start_date) & (df['transaction_date'] <= end_date)
df = df.loc[mask]
df.info()
```

## 4. Group a DataFrame by month and summarize the total amount for each month.

```python
import pandas as pd

# Sample data
data = {
    'transaction_id': [101, 102, 103, 104, 105, 106, 107],
    'transaction_date': ['2024-06-01', '2024-06-15', '2024-07-01', '2024-07-15', '2024-08-01', '2024-08-15', '2024-09-01'],
    'amount': [100, 200, 150, 300, 250, 350, 400]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df['transaction_date'] = pd.to_datetime(df['transaction_date'], format='%Y-%m-%d')
df['month'] = df['transaction_date'].dt.month
june_sum = 0
july_sum = 0
august_sum = 0
september_sum = 0
for _, row in df.iterrows():
    if row['month'] == 6:
        june_sum += row['amount']
    if row['month'] == 7:
        july_sum += row['amount']
    if row['month'] == 8:
        august_sum += row['amount']
    if row['month'] == 9:
        september_sum += row['amount']
data = {
    'month': ['2024-06','2024-07','2024-08','2024-09'],
    'amount': [june_sum, july_sum, august_sum, september_sum]
}
monthly_summary_df = pd.DataFrame(data)
monthly_summary_df


# Task: Convert 'transaction_date' to datetime, extract the month, and summarize the total amount for each month
```

## 6. Resample a DataFrame to aggregate data by month.

```python
import pandas as pd
# A dataframe with a DatetimeIndex is a time series dataframe, otherwise it is not
# when you do a group by, you have to specify a column name
# instead you can call the resample method on a time series dataframe
# ie. df.resample('ME').sum() # sum() is an example of an aggregation
# resampling is specific to data that has a DatetimeIndex

# Sample data
data = {
    'transaction_id': [101, 102, 103, 104, 105, 106, 107],
    'old_transaction_date': ['2024-06-01', '2024-06-15', '2024-07-01', '2024-07-15', '2024-08-01', '2024-08-15', '2024-09-01'],
    'amount': [100, 200, 150, 300, 250, 350, 400]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df['old_transaction_date'] = pd.to_datetime(df['old_transaction_date'])
df['transaction_date'] = df['old_transaction_date'].dt.month
df.drop('old_transaction_date', axis='columns', inplace = True)
gk = df.groupby('transaction_date').sum()
gk.rename(index={6: '2024-06-30', 7: '2024-07-31', 8: '2024-08-31', 9: '2024-09-30'}, inplace=True)
gk.reset_index(inplace=True)
gk['transaction_date'] = pd.to_datetime(gk['transaction_date'])
gk.set_index('transaction_date', inplace=True)
gk
```

```python


# Task: Convert 'transaction_date' to datetime, set it as index, and resample data to aggregate the total amount by month
```
