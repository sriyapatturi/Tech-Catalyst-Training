## Snowpark Usage - Example

````python
from snowflake.snowpark import Session
````

Example of how the `snow.cfg` file looks like


````python
import configparser

config = configparser.ConfigParser()
config.read('snow.cfg')
config.sections()
````

Create a dictionary (key-value) of the parameters

````python
params1 = dict(config['SNOW'])
params1
````

Passing the parameters as a dictionary to Session

````python
session1 = Session.builder.configs(params1).create()
````

````python
print(session1.get_current_user())
print(session1.get_current_database())
print(session1.get_current_schema())
````

Example using the `table` methods to call a specific table. In this example I am calling the `INS_ACCIDENTS` table

````python
accidents = session1.table("TECHCATALYST_DE.PUBLIC.INS_ACCIDENTS")
accidents.show(5)
````

Note, `accidents` is a Snowpark object (table). There is a difference between Snowpark DataFrames and Panda DataFrames. For now, you will convert to a Pandas DataFrame to do your transformations. Once done, you will convert back to a Snowpark DataFrame and write to the Database.

````python
accidents_df['ACCIDENT_DATE'] = pd.to_datetime(accidents_df['ACCIDENT_DATE'])
````

````python
accidents_df.info()
````

````python
accident_type = session1.table("TECHCATALYST_DE.PUBLIC.INS_ACCIDENT_TYPE")
accident_type_df = accident_type.to_pandas()
````

````python
accident_type_df.info()
````

````python
body_style= session1.table("TECHCATALYST_DE.PUBLIC.INS_BODY_STYLE")
body_style_df = body_style.to_pandas()
body_style_df.info()
````

````python
gender_marital_status = session1.table("TECHCATALYST_DE.PUBLIC.INS_GENDER_MARITAL_STATUS")
gender_martial_status_df = gender_marital_status.to_pandas()
gender_martial_status_df.info()
````

````python
insurance_coverage = session1.table("TECHCATALYST_DE.PUBLIC.INS_INSURANCE_COVERAGE")
insurance_coverage_df = insurance_coverage.to_pandas()
insurance_coverage_df.info()
````

````python
policyholder = session1.table("TECHCATALYST_DE.PUBLIC.INS_POLICYHOLDER")
policyholder_df = policyholder.to_pandas()
policyholder_df.info()
policyholder_df['BIRTHDATE'] = pd.to_datetime(policyholder_df['BIRTHDATE'])
````

````python
vehicles = session1.table("TECHCATALYST_DE.PUBLIC.INS_VEHICLES")
vehicles_df = vehicles.to_pandas()
vehicles_df.info()
````

````python
vehicle_use = session1.table("TECHCATALYST_DE.PUBLIC.INS_VEHICLE_USE")
vehicle_use_df = vehicle_use.to_pandas()
vehicle_use_df.info()
````

````python
# you want to use merge, not concat: 

accident_type_df.rename(columns={"ACCIDENT_TYPE_CODE": "ACCIDENT_TYPE", "ACCIDENT_TYPE": "ACCIDENT_TYPE_VARCHAR"})
````

````python
accidents_df.info()
````

````python
states_df.info()
````

````python
# FACT_ACCIDENTS
aph = pd.merge(accidents_df, policyholder_df, on = "POLICYHOLDER_ID", how = "inner")
aphat = pd.merge(aph, accident_type_df, on = "ACCIDENT_TYPE", how = "inner")
aphats = pd.merge(aphat, states_df, on = "STATE_CODE", how = "inner")
aphatsic = pd.merge(aphats, insurance_coverage_df, on = "POLICYHOLDER_ID", how = "inner")
aphatsicv = pd.merge(aphatsic, vehicles_df, on = "VEHICLE_ID", how = "inner")
aphatsicvbs = pd.merge(aphatsicv, body_style_df, on = "BODY_STYLE_CODE", how = "inner")
fact_accidents_df = pd.merge(aphatsicvbs, vehicle_use_df, on = "VEHICLE_ID", how = "inner")
fact_accidents_df.columns
fact_accidents_df = fact_accidents_df.drop(columns=['POLICYHOLDER_ID_y'])
fact_accidents_df = fact_accidents_df.rename(columns={"POLICYHOLDER_ID_x": "POLICYHOLDER_ID"})
fact_accidents_df = fact_accidents_df[["ACCIDENT_ID", "POLICYHOLDER_ID", "VEHICLE_ID", "STATE_CODE", "BODY_STYLE_CODE", "ACCIDENT_TYPE", "GENDER_MARITAL_STATUS", "USE_CODE", "ACCIDENT_DATE", "YEAR", "BIRTHDATE", "ESTIMATED_COST", "ACTUAL_REPAIR_COST", "AT_FAULT", "DUI", "COVERAGE_STATUS"]]
accidents_df

# "STATE_ID", "BODY_STYLE_ID", "ACCIDENT_TYPE_ID", "GENDER_MARITALSTATUS_ID", "VEHICLE_USECODE_ID", "ACCIDENT_DATE", "VEHICLE_YEAR", "POLICYHOLDER_BIRTHDATE", "ESTIMATED_COST", "ACTUAL_REPAIR_COST", "AT_FAULT", "IS_DUI", "COVERAGE_STATUS
````

````python
dim_accident_type_df = pd.merge(accidents_df, accident_type_df, on = "ACCIDENT_TYPE", how = "inner")
dim_accident_type_df = dim_accident_type_df[["ACCIDENT_ID", "ACCIDENT_TYPE_VARCHAR"]]
dim_accident_type_df
````

````python
dim_body_style_df = body_style_df[["BODY_STYLE_CODE", "BODY_STYLE"]]
dim_body_style_df
````

````python
dim_gender_marital_status_df = gender_martial_status_df[["GENDER_MARITAL_STATUS_CODE", "GENDER_MARITAL_STATUS"]]
dim_gender_marital_status_df
````

````python
dim_policyholder_df = policyholder_df[["POLICYHOLDER_ID", "FIRST_NAME", "LAST_NAME", "ADDRESS"]]
dim_policyholder_df
````

````python
dim_states_df = states_df[["STATE_CODE", "STATE"]]
dim_states_df
````

To avoid mistakes, make sure you use your schema.

Convert the Pandas DataFrame into a Snowpark DataFrame

````python
fact_accidents_sdf = session1.create_dataframe(fact_accidents_df)
dim_accident_type_sdf = session1.create_dataframe(dim_accident_type_df)
dim_body_style_sdf = session1.create_dataframe(dim_body_style_df)
dim_gender_marital_sdf = session1.create_dataframe(dim_gender_marital_status_df)
dim_policyholder_sdf = session1.create_dataframe(dim_policyholder_df)
dim_states_sdf = session1.create_dataframe(dim_states_df)

````

When using the `write.mode()` method there are different options:

* ”`append`”: Append data of this DataFrame to the existing table. Creates a table if it does not exist.

* ”`overwrite`”: Overwrite the existing table by dropping old table.

* ”`truncate`”: Overwrite the existing table by truncating old table.

* ”`errorifexists`”: Throw an exception if the table already exists.

* ”`ignore`”: Ignore this operation if the table already exists.

* Default value is “`errorifexists`”.
